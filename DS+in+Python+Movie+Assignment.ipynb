-  ### Subtask 2.3: Drop unecessary rows using columns with high Null percentages

Now, on inspection you might notice that some columns have large percentage (greater than 5%) of Null values. Drop all the rows which have Null values for such columns.

```python
# Write your code for dropping the rows here
null_percentages = (movies_cleaned.isnull().mean() * 100)

# Identify columns with more than 5% null values
columns_to_check = null_percentages[null_percentages > 5].index

# Drop rows with null values in these columns
movies_filtered = movies_cleaned.dropna(subset=columns_to_check)

print("DataFrame after dropping rows with null values:")
movies_filtered.head()
```

```python
print("\nShape of the filtered DataFrame:")
movies_filtered.shape
```

-  ### Subtask 2.4: Drop unecessary rows

Some of the rows might have greater than five NaN values. Such rows aren't of much use for the analysis and hence, should be removed.

```python
# Write your code for dropping the rows here
na_counts_per_row = movies_filtered.isnull().sum(axis=1)

# Filter out rows with more than five NaN values
movies_cleaned_final = movies_filtered[na_counts_per_row <= 5]

#inspect dataframe
print("DataFrame after removing rows with more than five NaN values:")
print(movies_cleaned_final.head())

print("\nShape of the final cleaned DataFrame:")
print(movies_cleaned_final.shape)

print("Remaining columns:")
print(movies_cleaned_final.columns)

print("\nShape of the final cleaned DataFrame:")
print(movies_cleaned_final.shape)
```

-  ### Subtask 2.5: Fill NaN values

You might notice that the `language` column has some NaN values. Here, on inspection, you will see that it is safe to replace all the missing values with `'English'`.

```python
# Write your code for filling the NaN values in the 'language' column here
movies_cleaned_final['language'] = movies_cleaned_final['language'].fillna('English')

print("DataFrame after replacing NaN values in 'language' with 'English':")
print(movies_cleaned_final.head())

print("\nNull values in 'language' column after replacement:")
print(movies_cleaned_final['language'].isnull().sum())
```

-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here
genres_split = movies_cleaned_final['genres'].str.split('|', expand=True)

# first and second genres to new columns
movies_cleaned_final['genre_1'] = genres_split[0]  # First genre
movies_cleaned_final['genre_2'] = genres_split[1].fillna(genres_split[0]) 

# Display the first rows
print(movies_cleaned_final[['genres', 'genre_1', 'genre_2']].head())
```

```python
movies_by_segment = movies_cleaned_final.groupby(['genre_1', 'genre_2'])

# Display the size of each group 
print(movies_by_segment.size().reset_index(name='count').head())
```

```python
grouped_genres = movies_cleaned_final.groupby(['genre_1', 'genre_2'])

# Mean gross value for each genre combination
mean_gross_by_genre = grouped_genres['gross'].mean().reset_index(name='mean_gross')

# Sort by mean gross value
mean_gross_by_genre_sorted = mean_gross_by_genre.sort_values(by='mean_gross', ascending=False)

PopGenre = mean_gross_by_genre_sorted.head(5)
print("Top 5 Popular Genre Combinations:")
print(PopGenre)
```

-  ### Subtask 3.7: Find the critic-favorite and audience-favorite actors

    1. Create three new dataframes namely, `Meryl_Streep`, `Leo_Caprio`, and `Brad_Pitt` which contain the movies in which the actors: 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' are the lead actors. Use only the `actor_1_name` column for extraction. Also, make sure that you use the names 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' for the said extraction.
    2. Append the rows of all these dataframes and store them in a new dataframe named `Combined`.
    3. Group the combined dataframe using the `actor_1_name` column.
    4. Find the mean of the `num_critic_for_reviews` and `num_user_for_review` and identify the actors which have the highest mean.

```python
# Write your code for creating three new dataframes here

Meryl_Streep = movies_cleaned_final[movies_cleaned_final['actor_1_name'] == 'Meryl Streep']
```

```python
Leo_Caprio = movies_cleaned_final[movies_cleaned_final['actor_1_name'] == 'Leonardo DiCaprio']
```

```python
Brad_Pitt = movies_cleaned_final[movies_cleaned_final['actor_1_name'] == 'Brad Pitt']
```

```python
# Write your code for combining the three dataframes here
Combined = pd.concat([Meryl_Streep, Leo_Caprio, Brad_Pitt], ignore_index=True)

print(Combined.head())
print(f"Total rows in Combined DataFrame: {len(Combined)}")
```

```python
# Write your code for grouping the combined dataframe here
grouped_by_actor = Combined.groupby('actor_1_name')

# Mean of 'num_critic_for_reviews' and 'num_user_for_reviews'
mean_reviews = grouped_by_actor[['num_critic_for_reviews', 'num_user_for_reviews']].mean().reset_index()

print("Mean Reviews by Actor:")
print(mean_reviews)
```

```python
# Write the code for finding the mean of critic reviews and audience reviews here
most_critic_favorite = mean_reviews.loc[mean_reviews['num_critic_for_reviews'].idxmax()]

# Actor with the highest mean audience reviews
most_audience_favorite = mean_reviews.loc[mean_reviews['num_user_for_reviews'].idxmax()]

# Results
print("Mean Reviews for Each Actor:")
print(mean_reviews)

print("\nMost Critic-Favorite Actor:")
print(most_critic_favorite)

print("\nMost Audience-Favorite Actor:")
print(most_audience_favorite)
```
